```python
# Función para imprimir una tabla de verdad de forma legible
def print_truth_table(title, headers, rows):
    print(f"\n--- {title} ---")
    print(" | ".join(headers))
    print("-" * (len(" | ".join(headers)) + 2))
    for row in rows:
        print(" | ".join([str(val).center(len(headers[i])) for i, val in enumerate(row)]))

# --- EJEMPLO 1: Implicación (Si llueve, entonces llevo paraguas) ---
# En lógica: L -> P es equivalente a (not L) or P
headers1 = ["L (Llueve)", "P (Paraguas)", "L -> P"]
rows1 = []
for L in [True, False]:
    for P in [True, False]:
        result = (not L) or P  # Equivalencia lógica de la implicación
        rows1.append([L, P, result])
print_truth_table("Implicación: L -> P", headers1, rows1)

# --- EJEMPLO 2: Disyunción (Voy al cine o me quedo en casa) ---
headers2 = ["C (Cine)", "H (Casa)", "C v H"]
rows2 = []
for C in [True, False]:
    for H in [True, False]:
        result = C or H
        rows2.append([C, H, result])
print_truth_table("Disyunción: C v H", headers2, rows2)

# --- EJEMPLO 3: Negación (No es cierto que me gusta la pizza) ---
headers3 = ["G (Gusta Pizza)", "not G"]
rows3 = []
for G in [True, False]:
    result = not G
    rows3.append([G, result])
print_truth_table("Negación: not G", headers3, rows3)

# --- EJEMPLO 4: Tautología (Llueve o no llueve) ---
# Siempre debe dar True (V)
headers4 = ["L (Llueve)", "not L", "L v (not L)"]
rows4 = []
for L in [True, False]:
    not_L = not L
    result = L or not_L
    rows4.append([L, not_L, result])
print_truth_table("Tautología: L v (not L)", headers4, rows4)

print("\nNota: En Python, 'True' representa 'V' (Verdadero) y 'False' representa 'F' (Falso).")
```
