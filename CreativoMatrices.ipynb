Para este ejemplo creativo, vamos a construir un "Explorador de Galaxias".

En lugar de ver una matriz como una simple cuadrícula de números, la trataremos como un mapa de coordenadas espaciales (x, y). Cada celda de la matriz representará un sector del espacio que puede contener un planeta, una estrella, un agujero negro o simplemente vacío.

Utilizaremos una matriz 2D (una lista de listas) para representar este universo.

```python
import random

# Definimos los símbolos de nuestra galaxia
VACIO = "."
ESTRELLA = "✨"
PLANETA = "🌍"
AGUJERO_NEGRO = "🕳️"

def crear_galaxia(filas, columnas):
    """Crea una matriz (galaxia) poblada aleatoriamente."""
    opciones = [VACIO] * 15 + [ESTRELLA] * 3 + [PLANETA] * 2 + [AGUJERO_NEGRO] * 1
    # Esta es la definición de la matriz: una lista de listas
    return [[random.choice(opciones) for _ in range(columnas)] for _ in range(filas)]

def mostrar_galaxia(galaxia):
    """Imprime la matriz de forma visual."""
    print("\n--- MAPA ESTELAR ---")
    for fila in galaxia:
        print(" ".join(fila))

def escanear_sector(galaxia, x, y):
    """Accede a un elemento específico de la matriz (coordenadas)."""
    if 0 <= x < len(galaxia) and 0 <= y < len(galaxia[0]):
        objeto = galaxia[x][y]
        mensajes = {
            VACIO: "Espacio profundo... nada a la vista.",
            ESTRELLA: "¡Cuidado! Una estrella masiva detectada.",
            PLANETA: "¡Señal de vida! Un planeta habitable encontrado.",
            AGUJERO_NEGRO: "¡ALERTA! Horizonte de sucesos detectado. ¡Huye!"
        }
        return mensajes.get(objeto, "Error de escaneo.")
    return "Coordenadas fuera de la galaxia conocida."

# --- EJECUCIÓN CREATIVA ---

# 1. Definimos las dimensiones de nuestra matriz (10x10)
ANCHO, ALTO = 10, 10
mi_galaxia = crear_galaxia(ALTO, ANCHO)

# 2. Mostramos el universo visualmente
mostrar_galaxia(mi_galaxia)

# 3. Simulamos un escaneo en una coordenada específica (fila 3, columna 5)
coord_x, coord_y = 3, 5
resultado = escanear_sector(mi_galaxia, coord_x, coord_y)

print(f"\nEscaneando coordenadas ({coord_x}, {coord_y})...")
print(f"Resultado del radar: {resultado}")

# 4. Operación de matriz: Contar objetos totales
total_objetos = sum(1 for fila in mi_galaxia for celda in fila if celda != VACIO)
print(f"\nTotal de anomalías detectadas en esta galaxia: {total_objetos}")
```

¿Por qué este código es un gran ejemplo de matrices?

1. Estructura de datos: mi_galaxia es una lista de listas. Para acceder a un punto específico, usamos galaxia[x][y]. El primer índice (x) elige la fila y el segundo (y) la columna.
2. Recorrido de matriz: La función mostrar_galaxia recorre la matriz fila por fila para imprimirla, lo cual es la operación más básica y esencial al trabajar con datos tabulares.
3. Comprensión de listas anidadas: La línea sum(1 for fila in mi_galaxia for celda in fila if celda != VACIO) es una forma muy elegante y potente de "aplanar" la matriz mentalmente para contar elementos, algo que se usa constantemente en análisis de datos e IA.
4. Coordenadas vs Índices: Este código ayuda a entender que en programación de matrices, la ubicación física de un dato es tan importante como el valor del dato mismo.

Dato extra para IA: En el mundo del procesamiento de imágenes, las fotos no son más que matrices gigantes donde cada celda es un píxel con un valor de color. ¡Este script es básicamente el principio de cómo una IA "ve" una imagen!
