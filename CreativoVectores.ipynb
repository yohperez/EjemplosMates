Para este ejemplo creativo, vamos a crear un "Sistema de Clasificación de Superhéroes". En lugar de aburridos números, vamos a ver cómo cada héroe es un vector de personalidad en un espacio de 3 dimensiones: Fuerza (F), Agilidad (A) e Inteligencia (I).

Este script calculará quién es el superhéroe más parecido a otro basándose en la distancia euclidiana (una forma fundamental en que la IA entiende la similitud).

```python
import math

# Definimos a nuestros héroes como vectores: [Fuerza, Agilidad, Inteligencia]
# Escala del 0 al 10
heroes = {
    "Hulk":       [10, 2, 1],
    "Spider-Man": [4, 10, 7],
    "Iron Man":   [7, 5, 10],
    "Thor":       [9, 4, 6],
    "Black Widow":[5, 9, 8],
    "Captain Am.":[7, 7, 7]
}

def calcular_distancia(vec1, vec2):
    """Calcula la distancia euclidiana entre dos vectores."""
    # La fórmula es la raíz cuadrada de la suma de las diferencias al cuadrado
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(vec1, vec2)))

def encontrar_aliado(nombre_hero):
    if nombre_hero not in heroes:
        return "Héroe no encontrado."

    vector_objetivo = heroes[nombre_hero]
    mejor_match = None
    distancia_minima = float('inf')

    print(f"Buscando el aliado más parecido a {nombre_hero}...")

    for nombre, vector in heroes.items():
        if nombre == nombre_hero:
            continue

        distancia = calcular_distancia(vector_objetivo, vector)
        if distancia < distancia_minima:
            distancia_minima = distancia
            mejor_match = nombre

    return mejor_match, distancia_minima

# --- EJECUCIÓN CREATIVA ---
hero_elegido = "Spider-Man"
aliado, distancia = encontrar_aliado(hero_elegido)

print(f"El aliado más cercano a {hero_elegido} es {aliado} (Distancia: {distancia:.2f})")

# Visualización de cómo la IA "ve" los vectores
print("\n--- Vista de Vectores (Nivel de IA) ---")
for nombre, v in heroes.items():
    print(f"{nombre}: {v}")
```

¿Por qué es creativo este código?

1. Representación espacial: Al definir a los héroes como una lista de tres números [F, A, I], los estamos convirtiendo en puntos en un espacio 3D. La IA "entiende" que si dos puntos están cerca, los héroes son similares.
2. La métrica de distancia: Usamos math.sqrt y una comprensión de listas para calcular la diferencia. Esta es la misma lógica que utilizan algoritmos de aprendizaje automático como K-Nearest Neighbors (KNN) para clasificar datos.
3. Modularidad: Puedes añadir más héroes al diccionario heroes y el código funcionará inmediatamente sin cambiar una sola línea de lógica.

 Si añado una cuarta dimensión al vector (por ejemplo, [Fuerza, Agilidad, Inteligencia, Dinero]), el código funcionará igual y el cálculo será más preciso, aunque ya no podríamos dibujarlo en un plano 3D. Si comparamos a Batman (quien tendría un valor muy alto en la dimensión de "Dinero") la comparacion va a ser muy desigual
