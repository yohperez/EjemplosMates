```python
# Base de datos de películas y sus etiquetas (keywords)
# Usamos sets para una búsqueda y comparación O(1) ultra rápida
movies_db = {
    "Inception": {"sci-fi", "dream", "mind-bending", "action", "heist"},
    "Interstellar": {"sci-fi", "space", "time-travel", "emotion", "adventure"},
    "The Dark Knight": {"action", "crime", "hero", "dark", "justice"},
    "Pulp Fiction": {"crime", "nonlinear", "dialogue", "dark"},
    "The Matrix": {"sci-fi", "action", "simulation", "future"},
    "Toy Story": {"animation", "comedy", "adventure", "friendship"},
    "Finding Nemo": {"animation", "adventure", "ocean", "family"},
    "Joker": {"crime", "drama", "psychological", "dark"},
    "Avatar": {"sci-fi", "fantasy", "adventure", "future"},
    "Arrival": {"sci-fi", "aliens", "linguistics", "time-travel"},
    "Gladiator": {"action", "history", "war", "drama"}
}

def get_all_keywords(db):
    """Crea el vocabulario universal de etiquetas"""
    return sorted(set().union(*db.values()))

def create_vector(movie_keywords, all_keywords):
    """Crea un vector binario (1 si tiene la etiqueta, 0 si no)"""
    return [1 if kw in movie_keywords else 0 for kw in all_keywords]

def jaccard_similarity(set1, set2):
    """Calcula la similitud entre 0 y 1"""
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

def search_movies_by_keyword(keyword):
    """Busca películas que contienen una palabra clave"""
    found = [title for title, tags in movies_db.items() if keyword.lower() in tags]
    return found

# --- Lógica principal ---

# 1. Crear vectores
all_keywords = get_all_keywords(movies_db)
print(f"Vocabulario total ({len(all_keywords)} etiquetas): {', '.join(all_keywords)}")

# 2. Comparar dos películas
movie1, movie2 = "Inception", "The Matrix"
sim = jaccard_similarity(movies_db[movie1], movies_db[movie2])
print(f"\nSimilitud entre {movie1} y {movie2}: {sim:.2%}")

# 3. Mostrar vectores (representación numérica)
print(f"\nVector de '{movie1}': {create_vector(movies_db[movie1], all_keywords)}")
print(f"Vector de '{movie2}': {create_vector(movies_db[movie2], all_keywords)}")

# 4. Búsqueda por palabra clave
search_term = "sci-fi"
results = search_movies_by_keyword(search_term)
print(f"\nPelículas con la etiqueta '{search_term}': {', '.join(results)}")
```
