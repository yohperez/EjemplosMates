Imagina que un grupo de amigos quiere saber mas sobre su cosecha de Manzanas.
Cosechan 10 manzanas y quieren analizar su peso para ver como es su producción este año.
Queremos calcular:
Media(promedio) Mediana(peso central) Moda(Peso mas común) Desviación Estandar(Qué tanto varian los pesos)

```python
import numpy as np
import statistics as stats
manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

#Estadísticas descriptivas
media = np.mean(manzanas) #promedio
mediana = np.median(manzanas) #valor central
moda = stats.mode(manzanas) #valor más común
desviación = np.std(manzanas) #variabilidad
#Resultados
print(f"Análisis de la cosecha de manzanas 🍎")
print(f"Media: {media} gramos")
print(f"Mediana: {mediana} gramos")
print(f"Moda: {moda} gramos")
print(f"Desviación Estandar: {desviación} gramos")
```
